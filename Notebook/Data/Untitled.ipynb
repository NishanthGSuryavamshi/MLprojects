```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re

# Load the Excel file
df = pd.read_excel(r'C:\Users\vamsh\Downloads\20211030 Test Assignment-20230411T191020Z-001\20211030 Test Assignment\Input.xlsx')

# Define a function to extract the article text
def extract_article(link):
    # Send a request to the website
    response = requests.get(link)
    # Parse the HTML content of the website
    soup = BeautifulSoup(response.content, 'html.parser')
    # Find the article text
    article = soup.find('article')
    # Extract the title and text from the article
    title = article.h1.string
    text = article.get_text()
    # Remove unnecessary text
    text = re.sub(r'\n\s*\n', '\n', text)
    text = re.sub(r'\n+', '\n', text)
    text = re.sub(r'^\s+|\s+$', '', text)
    # Create a file with the URL_ID as its name and write the extracted text to the file
    with open(str(df[df['Link'] == link]['URL_ID'].values[0]) + '.txt', 'w') as file:
        file.write(title + '\n' + text)

# Apply the function to the links in the Excel sheet
df['Link'].apply(extract_article)
```

```python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import cmudict, stopwords

# load resources for syllable count
nltk.download('cmudict')
nltk.download('stopwords')
d = cmudict.dict()
stop_words = set(stopwords.words('english'))

# function to calculate positive score
def get_positive_score(text):
    positive_words = set(nltk.corpus.opinion_lexicon.positive())
    tokens = nltk.word_tokenize(text.lower())
    return len([t for t in tokens if t in positive_words])

# function to calculate negative score
def get_negative_score(text):
    negative_words = set(nltk.corpus.opinion_lexicon.negative())
    tokens = nltk.word_tokenize(text.lower())
    return len([t for t in tokens if t in negative_words])

# function to calculate polarity score
def get_polarity_score(text):
    positive_score = get_positive_score(text)
    negative_score = get_negative_score(text)
    return (positive_score - negative_score) / (positive_score + negative_score + 1e-9)

# function to calculate subjectivity score
def get_subjectivity_score(text):
    return nltk.sentiment.util.extract_unigram_feats(nltk.sentiment.util.mark_negation(nltk.word_tokenize(text.lower()))).get('subj', 0)

# function to calculate average sentence length
def get_avg_sentence_length(text):
    sentences = sent_tokenize(text)
    return sum(len(word_tokenize(s)) for s in sentences) / len(sentences)

# function to calculate percentage of complex words
def get_percentage_complex_words(text):
    tokens = word_tokenize(text.lower())
    num_complex = len([t for t in tokens if len(t) > 2 and t not in stop_words and len(d.get(t, [])) > 1])
    return num_complex / len(tokens)

# function to calculate FOG index
def get_fog_index(text):
    num_sentences = len(sent_tokenize(text))
    num_words = len(word_tokenize(text.lower()))
    num_complex = len([t for t in word_tokenize(text.lower()) if len(t) > 2 and t not in stop_words and len(d.get(t, [])) > 1])
    return 0.4 * ((num_words / num_sentences) + (num_complex / num_words) * 100)

# function to calculate average number of words per sentence
def get_avg_words_per_sentence(text):
    sentences = sent_tokenize(text)
    return sum(len(word_tokenize(s)) for s in sentences) / len(sentences)

# function to calculate complex word count
def get_complex_word_count(text):
    tokens = word_tokenize(text.lower())
    return len([t for t in tokens if len(t) > 2 and t not in stop_words and len(d.get(t, [])) > 1])

# function to calculate word count
def get_word_count(text):
    tokens = word_tokenize(text.lower())
    return len(tokens)

# function to calculate syllable per word
def get_syllable_per_word(text):
    tokens = word_tokenize(text.lower())
    num_syllables = sum([len(d.get(t, [])) for t in tokens])
    return num_syllables / len(tokens)

# function to calculate personal pronouns
def get_personal_pronouns(text):
    personal_pronouns = ['i', 'me', 'my', 'mine', 'we', 'us', 'our', 'ours']
    tokens = nltk.word_tokenize(text.lower())
    return
```
